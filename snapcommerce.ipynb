```python
#Setup

##Download data into dataframes
sheet_id = "1TDJdzeS7VdiR4ZsmvGSV9QtgsUY6Rv5oQlmyw59Bftw"
sheet_names = ["customer_data", "order_data", "order_line_data", "product_data"]
tables = {}

url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet="


for sheet_name in sheet_names:
    tables[sheet_name] = pd.read_csv(url+sheet_name)
    
##Standardize all date columns to date time object since I noticed inconsistent dates
tables['customer_data']['created_at'] = pd.to_datetime(tables['order_data']['created_at'])
tables['order_data']['created_at'] = pd.to_datetime(tables['order_data']['created_at'])
tables['order_data']['refunded_at'] = pd.to_datetime(tables['order_data']['refunded_at'])
tables['product_data']['created_at'] = pd.to_datetime(tables['product_data']['created_at'])

##Extract from dictionary into individual variables for querying
customer_data, order_data, order_line_data, product_data = [tables.get(t) for t in sheet_names]
```

```python
###########
#Question 1:
###########


##All pretty simple queries, so didn't change the data model

##Gross Merchandise Value
####Total price of all non refunded orders
psql.sqldf("""
select sum(total_price)
from order_data
where refunded_at is NULL
""")

##Average Order Value
####Total Price / number of orders, will include refunded orders
psql.sqldf("""
select sum(total_price) / count(*)
from order_data
--where refunded_at is NULL
""")

##Average Basket Size
####Interpreting this as average quantity per order
psql.sqldf("""
select avg(order_quantity)
from
(
select order_id, sum(quantity) order_quantity
from order_line_data
group by 1
)
""")

##Monthly Active Customers
####For each month, the number of customers who made a purchase
psql.sqldf("""
select strftime("%m-%Y", created_at) as 'month-year' , count(*)
from order_data
group by 1
""")
```

```python
###########
#Question 2:
###########


###NOTES###

##Customer retention rate would probably be easier to calculate if we had a customer_retention table
##Keep granularity at the user-order level so we can join to various dimensions
##Split date parts out so we could do accurate and scalable month retention (not all months have 30 days, etc)
##Most analytical data warehouses have datediff functions which would make our lives easier

##Do NOT bring dimensions into customer_retention table since dimensions will likely scale
##and don't want to edit this table every time we have a new dimension
##Tradeoff is we will constantly be joining back to primary tables, which could result in slower queries, but best
##practice is to not have multiple places to change code

##With a small amount of data, recreating the customer_retention table each night is not expensive
##With a larger set of data, and if we defined which retention constraints we wanted, we should only look back in
##the past year/month/day rather than recreate entire table each night


customer_retention = psql.sqldf("""
select c.id customer_id, 
o.id order_id, 
c.created_at registration_date, 
o.created_at order_placed_date,
first_order_id,
CAST(strftime('%Y', c.created_at) as INT) registration_year, 
CAST(strftime('%m', c.created_at) as INT) registration_month, 
CAST(strftime('%d', c.created_at) as INT) registration_day, 

CAST(strftime('%Y', o.created_at) as INT) order_placed_year, 
CAST(strftime('%m', o.created_at) as INT) order_placed_month, 
CAST(strftime('%d', o.created_at) as INT) order_placed_day
from customer_data c
left join order_data o on c.id = o.customer_id
join (
select customer_id, 
first_value(id) over (partition by customer_id order by created_at) first_order_id
from order_data
group by 1
) first_order on c.id = first_order.customer_id
""")
```

```python
##Below are queries which result in a basic retention table
##Did monthly retention, but easily change if we wanted yearly or annual
##From there, if we have some sort of BI tool (Looker, Tableau, etc) could easily put that on top for visualizations
##which would automatically update each time the Customer Retention ETL was run

#Basic Customer Retention table
psql.sqldf("""
select 
strftime('%Y %m', registration_date) reg_cohort_month,
(order_placed_year - registration_year) * 12 + (order_placed_month - registration_month) month,
count(distinct(customer_id)) num_customers
FROM customer_retention
group by 1,2
""")

#Customer Retention split by dimensions (eg gender)
psql.sqldf("""
select gender,
strftime('%Y %m', registration_date) reg_cohort_month,
(order_placed_year - registration_year) * 12 + (order_placed_month - registration_month) month,
count(distinct(customer_id)) num_customers
FROM customer_retention cr
join customer_data cd on cr.customer_id = cd.id
group by 1,2,3
""")

#Customer Retention split by first order attributes (eg product category)
psql.sqldf("""
select 
pd.category,
strftime('%Y %m', registration_date) reg_cohort_month,
(order_placed_year - registration_year) * 12 + (order_placed_month - registration_month) month,
count(distinct(customer_id)) num_customers
FROM customer_retention cr
left join order_line_data old on cr.first_order_id = old.order_id
join product_data pd on pd.id = old.product_id
group by 1,2,3
""")
```

```python
###Customer Lifetime Value
###Different definitions for CLV but usually average total value of a customer (sometimes timeboxed)
###The order_data table has all the necessary information
###Will ignore refunded_at, but could easily filter out if we wanted to

psql.sqldf("""
select avg(customer_total_value)
from
(
select customer_id, sum(total_price) customer_total_value
from order_data
group by 1
)
""")

##If we wanted a timebox version of this (eg CLV for 1 month period) we would filter to only customers who registered in January
##1 Month CLV (average amount we expect a customer to bring in in their 1st month after registering)
##We could recycle our customer_retention table since it has all the necessary attributes
psql.sqldf("""
select avg(customer_value_month1)
from
(
select cr.customer_id, sum(total_price) customer_value_month1
from customer_retention cr
join order_data od on od.id = cr.order_id
where registration_year = order_placed_year
and registration_month = order_placed_month
group by 1
)
""")
```
